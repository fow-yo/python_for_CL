```python
import requests as rq

from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://4stor.ru/'
```

```python
page = rq.get(url)
page
```

```python
page.text
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
print(soup.prettify())
```

```python
data={}
stories = soup.find_all('section', class_="story_item")
for i in range(len(stories)):
    item=stories[i]
    titledata={}
    titledata['rate'] = item.find(class_='ratingtypeplusminus').text
    titledata['title'] =item.h2.span.text
    titledata['descrip']= item.find(class_='desc').p.text
    titledata['author']= item.find(class_='autor').text
    titledata['info']= item.find(class_='info').text
    data.update({i: titledata})

data
    

```
